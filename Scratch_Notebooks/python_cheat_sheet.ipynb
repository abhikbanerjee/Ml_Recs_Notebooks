## Use np.Vectorize to return functions

```python
@np.vectorize
def noneg(n):
    if not np.isnan(n) and n <0:
        return n.__class__(-1)
    else:
        return n
```

```python
v = np.array([-1,3,5,-33,2])
noneg(-3)
noneg(v)
```

```python
v1 = np.array([-1, 2, np.nan])
```

```python
noneg(v1)
```
