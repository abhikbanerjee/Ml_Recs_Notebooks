```python
import pandas as pd

# Read the olympics data
oo = pd.read_csv("./olympics.csv", skiprows=4)
```

```python
oo.head()
```

### Basic usages of value_counts() , sort_values() and boolean operations

```python
print(type(oo.City))

# Use of value counts
oo.Gender.value_counts(ascending=True)
```

```python
oo.info()
```

```python
# use of sort_values
oo.sort_values(by=['City', 'Edition'], ascending=[True, False]).head(5)
```

```python
# Use of boolean conditions and str operations
ma = oo[(oo.Gender == 'Men') & (oo.Sport == 'Aquatics')]
ma[ma.Athlete.str.contains('Otto')]

```
