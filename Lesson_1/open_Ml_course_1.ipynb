## Read a data into a Pandas DF

## Use Describe, Value Counts to check the various kind of variables and values

```python
df.describe() # include can be passed to include object and bool values
print("******")
df.describe(include=['object','bool','int64','float64'])
```

```python
## Convert a column type using astype()

df["Churn"] = df.Churn.astype('int64')
print(df.head())
```

```python
# do a value counts and proportions for the categorical columns
df['Churn'].value_counts()

#Use normalize=True to get the proportions
df['Churn'].value_counts(normalize=True)
```

## Some quick queries and operations on Pandas

### What are average values of numerical variables for churned users?

```python
print(df[df["Churn"]==1].mean())
```

### How much time (on average) do churned users spend on phone during daytime?

```python
print(df[df["Churn"]==1]["Total day minutes"].mean())
```

### What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?

```python
print(df[ (df['Churn']==0) & (df['International plan']=="No")]["Total intl minutes"].max())
```


## Applying functions to Rows, Columns and Cells

```python
# Apply can be used to apply a function to the dataframe
import numpy as np

df.apply(np.max)
```

### Find all the states that start with a W or C

```python
df[df["State"].apply(lambda state : (state[0]=="W" or state[0]=="C"))].head(n=12)
```

### Use Replace or map function to replace values

```python
## use map function to replace some of the values

d = {"No":False, "Yes":True}
df["International plan"] = df["International plan"].map(d)
```

```python
# Also can be done the same thing using replace command
df = df.replace({"Voice mail plan" : d})
df.head()
```

## Group by Clauses

```python
cols_to_show = ['Total day minutes','Total day calls', 'Total day charge']
# df.groupby('Churn')[cols_to_show].describe()
df.groupby('Churn')[cols_to_show].describe(percentiles=[])
```

```python
df.groupby('Churn')[cols_to_show].agg(['mean','std','min','max'])
# df.groupby('Churn')[cols_to_show].agg([np.mean,np.std,np.min,np.max])   # can be also used
```

## Summary Tables usage and Pivot tables 

```python
# Use cross tab
pd.crosstab(df["Churn"], df["International plan"])
```

```python
pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
```

```python
# Use Pivot table
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'], ['Area code'], aggfunc='mean')
```

```python
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'], ['State'], aggfunc='mean').head(n=15)
```

## DataFrame Transformations

```python
df['Total Spent Mins'] = df["Total day minutes"]+df["Total eve minutes"]+ \
                            df["Total night minutes"]+df["Total intl minutes"]
    
df.head()
```

```python
df.drop(['Total Spent Mins'], axis=1, inplace=True)
df.head()
```

## Predicting Telecom Churn

```python
# Define the figuresize and some helpful imports
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rcParams['figure.figsize']=(8,6)
```

```python
pd.crosstab(df['International plan'], df['Churn'], margins=True)
```

```python
sns.countplot(x='International plan', hue='Churn', data=df)
```

```python
pd.crosstab(df['Customer service calls'], df['Churn'], margins=True)
```

```python
sns.countplot(x='Customer service calls', hue='Churn', data=df)
```

```python
df["Many Service Calls"] = (df['Customer service calls']>3)
# df["Many Service Calls"] = (df['Customer service calls']>3).astype(int)
pd.crosstab(df['Many Service Calls'], df['Churn'], margins=True)
```

```python
sns.countplot(x='Many Service Calls', hue='Churn', data=df)
```

```python
pd.crosstab((df['International plan'] & df['Many Service Calls']), df['Churn'], margins=True)
```
