### Plot  a histogram on Total day minutes and Total Intl calls

```python
features = ['Total day minutes','Total intl calls']
df[features].hist(figsize=(12,4))
```

```python
df[features].plot(kind='density', subplots=True, figsize=(12,4), layout=(1,2), sharex=False)
```

```python
# Display seaborn's distplot for one of the variables

sns.distplot(df['Total intl calls'])
```

```python
# Make a box plot using seaborn
_,ax = plt.subplots(figsize=(3,4))
sns.boxplot(data=df['Total intl calls'], ax=ax)
```

```python
_, ax = plt.subplots(1,2, sharey=True, figsize=(6,4))
sns.boxplot(data= df['Total intl calls'], ax=ax[0])                    
sns.violinplot(data = df['Total intl calls'], ax=ax[1])                    
```

```python
import numpy as np
# plot a sigmoid function
x = np.arange(-20,20,0.1)
y = 1/(1+np.exp(-x))
```

```python
import math

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
plt.plot(x,sig)
plt.show()
```
